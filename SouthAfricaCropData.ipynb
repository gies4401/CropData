```python
#ALL SOUTH AFRICA DATA FOUND FROM https://www.sagis.org.za/cec_reports.html
#By clicking each year on the link, I was able to find the final crop data from the year before. Each year is presented as the 1998-1999 growing season (as an example).
#This data goes back to the 1998-1999 growing season.
#Within each year in the link, I found the most recent winter/summer data from the year before; it was most likely presented as "Final Crop data for year ___", but ocassionally, final data was represented or had a broken link so I had to grab the most recent data estimates/production forecast, which I found to have very similar numbers as the final crop data would've had.
#I went through each year of data and had to create separate excel files for each season, year, and 9 given locations' crop data. I collected data of the area planted and converted that number from hectares to acres, and I also collected the final amount of crop harvested in tonnes and placed that data in the excel sheet.
#I collected data for winter wheat and summer corn and soybeans.
#Using python, I have all the national data from each year and each seaosn collected and compiled into csv files, one for each corn, soy, and wheat.
#I have also created separate csv files for the data of each of the 9 locations from every year and did it for every location of every season and crop.
#Finally, I combined all the data for each crop (locations and years) into their own csv files (totalling 3 files for corn, soy, and wheat)
#The order of this code series starts with the compiled national data and then goes into each csv file for each location of each crop, and it finishes with the csv files of the combined location and year crop data.

#National South Africa Winter Wheat Data
import pandas
import glob
#these imports allow me to call and use files stored on my computer.
directory = r'C:\Users\Sydney\Documents\2020 Internship\South Africa\South Africa Winter National'
files = glob.glob(directory + "/*.xlsx")
fields = ["Area Planted (acres)", "Harvested (tons)"]
records = []

for f in files:
    # Every file is put in a dataframe, and operations are performed
    df = pandas.read_excel(f, skipinitialspace=True, header=1, usecols=fields)
    max_value = df["Area Planted (acres)"].max()
    max2_value = df["Harvested (tons)"].max()
    file_name = f.replace(directory, '')
    file_name = file_name.replace('.xlsx', '')
    file_name = file_name.replace('\\', '')
    records.append((file_name, max_value, max2_value))

# Finally, we create a dataframe with our records and we store it
df = pandas.DataFrame.from_records(records, columns=['Year', 'Area Planted (acres)', 'Harvested (tonnes)'])
df.to_csv('SAWinter_National.csv', sep="," , index=False)
```

```python
#All Summer Soybean data from the 9 locations across South Africa 
import pandas
import glob
directory = r'C:\Users\Sydney\Documents\2020 Internship\South Africa\South Africa Summer Provinces'
fields = ["Area Planted (acres)", "Harvested (tons)"]
records = []

for f in glob.glob(directory + "/*/*.xlsx"):
    # Every file is put in a dataframe, and operations are performed
    df = pandas.read_excel(f, skiprows=3, header=1, usecols=fields)
    max_value = df["Area Planted (acres)"].max()
    max2_value = df["Harvested (tons)"].max()
    file_name = f.replace(directory, '')
    file_name = file_name.replace('.xlsx', '')
    file_name = file_name.replace('\\', '')
    f_length = len(file_name)
    f_length2 = f_length - 9
    location = file_name[0:f_length2]
    year = file_name[f_length2:f_length]
    records.append((location, year, max_value, max2_value,))

# Finally, we create a dataframe with our records and we store it
df = pandas.DataFrame.from_records(records, columns=['Location', 'Year', 'Area Planted (acres)', 'Harvested (tonnes)'])
df.to_csv('SASummer_Soy.csv', sep="," , index=False)
```
